```python
import numpy as np

def compare_priorities(priorities1, priorities2, label1, label2):
    # Sort priorities1 by value in descending order
    sorted_priorities1 = dict(sorted(priorities1.items(), key=lambda x: x[1], reverse=True))

    # Normalize the weights in both priorities
    total1 = sum(sorted_priorities1.values())
    sorted_priorities1 = {k: v/total1*100 for k,v in sorted_priorities1.items()}
    
    total2 = sum(priorities2.values())
    priorities2 = {k: v/total2*100 for k,v in priorities2.items()}

    # Get weights from priorities2 for matching keys in sorted_priorities1
    weights1 = list(sorted_priorities1.values())
    weights2 = [priorities2.get(goal, 0) for goal in sorted_priorities1.keys()]

    # Positions for each bar
    bar_width = 0.35
    r1 = np.arange(len(sorted_priorities1))
    r2 = [x + bar_width for x in r1]

    # Create bar graph
    plt.bar(r1, weights1, color='b', width=bar_width, edgecolor='grey', label=label1)
    plt.bar(r2, weights2, color='r', width=bar_width, edgecolor='grey', label=label2)
    plt.xlabel('Career Goals')
    plt.ylabel('Relative Importance (%)')
    plt.title('Comparison of Career Goals Priorities')
    plt.xticks([r + bar_width/2 for r in range(len(sorted_priorities1))], sorted_priorities1.keys(), rotation=45)
    plt.legend()

    plt.show()

def calculate_distance(priorities1, priorities2):
    # Get all keys
    all_keys = set(list(priorities1.keys()) + list(priorities2.keys()))

    # Calculate total distance
    total_distance = sum(abs(priorities1.get(key, 0) - priorities2.get(key, 0)) for key in all_keys)

    return total_distance


# Define priorities for comparison
goals = {'Compensation': 10, 'Daily Interest': 9, 'Learning':7, 'Industry': 5,'Location': 5, 'Work/Life': 5,'Growth': 5,'Culture': 8}
opportunity = {'Compensation': 8, 'Daily Interest': 7, 'Learning': 8, 'Industry': 8,'Location': 9, 'Work/Life': 7,'Growth': 6, 'Culture': 8}

# Call the function
compare_priorities(goals, opportunity, 'Goals', 'Quizlet')


# Call the function with your priorities
distance = calculate_distance(goals, opportunity)
print(f"The Manhattan distance between the two sets of priorities is: {distance}")
```
